### Connection set-up:

```python
import torch, os
from google.colab import (drive, files)

PATH = "/content/drive"

match os.path.exists(PATH):
    # Check for Duplicate Mount:
    case True:
        print("Drive is already mounted.")
    case False:
        drive.mount(PATH)
        files.upload()



    # GPU Resource Check:
        match torch.cuda.is_available():
            case True:
                print(f"{torch.cuda.device_count()} GPU(s) are available.")
            case False:
                print("No GPU. Using default resources.")
                print(f"Version: {torch.__version__}")
```
